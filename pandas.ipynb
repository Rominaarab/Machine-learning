```python
import pandas as pd
import numpy as np

df_demo = pd.DataFrame ({
    "A" : 1.2,
    "B" : pd.Timestamp('20180226'),
    "C" : [(-1)**i * np.sqrt(i) + np.e * (-1) ** (i-1) for i in range (5)],
    "D" : pd.Categorical(["This", "column", "has", "entries", "entries"]),
    "E" : "Same"
})
df_demo
```
